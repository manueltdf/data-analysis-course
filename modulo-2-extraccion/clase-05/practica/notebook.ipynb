### Ejercicio 1 — Una Series y su diferencia con una lista
 
Antes de entender por qué Pandas es útil, hay que sentir en carne propia qué pasa con una lista común cuando intentás hacer lo mismo que con una Series.
 
**Consigna:** Primero, ejecuta este código tal como está y observa qué pasa:
 

````python
edades = [22, 35, 28, 41, 19]
edades_en_5_anios = edades + 5
print(edades_en_5_anios)
````

Anota el error exacto que obtienes. Ahora convierte `edades` en una Series de Pandas y repite la misma operación.

````python
import pandas as pd
 
edades = pd.Series([22, 35, 28, 41, 19])
 
# Tu tarea: sumarle 5 a cada edad, sin usar un bucle
edades_en_5_anios = edades + 5
 
print(edades_en_5_anios)
````

### Ejercicio 4 — El error de un solo corchete
 
Este es uno de los errores más comunes al empezar con Pandas, y entenderlo bien te ahorra mucha confusión más adelante.
 
**Consigna:** Ejecuta cada una de estas líneas por separado y anota qué tipo de objeto devuelve cada una (puedes usar `type()` para verificarlo). Presta atención a la diferencia entre usar un corchete y dos.


````python
import pandas as pd
 
productos = pd.DataFrame([
    {"nombre": "Notebook", "precio": 320000, "stock": 5},
    {"nombre": "Mouse",    "precio": 8500,   "stock": 42},
    {"nombre": "Teclado",  "precio": 45000,  "stock": 12},
])
 
resultado_a = productos["precio"]
resultado_b = productos[["precio"]]
resultado_c = productos[["nombre", "precio"]]
 
print(type(resultado_a))
print(type(resultado_b))
print(type(resultado_c))
````

## Ejercicio 5 — Seleccionar filas con iloc
 
Un profesor tiene las notas de un parcial cargadas en un DataFrame y necesita revisar casos puntuales: el primer alumno de la lista, los últimos tres registros, y un rango específico para una segunda corrección.
 
**Consigna:** Usando `.iloc[]`, resuelve cada consulta:

````python
import pandas as pd
 
notas = pd.DataFrame([
    {"alumno": "Ana",    "nota": 8.5},
    {"alumno": "Luis",   "nota": 6.0},
    {"alumno": "Clara",  "nota": 9.2},
    {"alumno": "Pedro",  "nota": 5.5},
    {"alumno": "María",  "nota": 7.8},
    {"alumno": "Jorge",  "nota": 4.0},
])
 
# 1. Muestra solo la primera fila
primera_fila = notas.iloc[0]
 
# 2. Muestra las filas en las posiciones 2, 3 y 4 (Clara, Pedro, María)
rango_medio = notas.iloc[2:5]
# 3. Muestra solo la última fila
# Pista: pensá en los índices negativos que usaste con listas en clases anteriores
ultima_fila = notas.iloc[-1]
 
print(primera_fila)
print(rango_medio)
print(ultima_fila)
````
